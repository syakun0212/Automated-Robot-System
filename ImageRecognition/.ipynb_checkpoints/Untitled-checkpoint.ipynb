```python
import cv2
import numpy as np

def find_centroids(dst):
    ret, dst = cv2.threshold(dst, 0.01 * dst.max(), 255, 0)
    #th2 = cv2.adaptiveThreshold(dst, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
    cv2.imshow("AT", dst)
    dst = np.uint8(dst)

    # find centroids
    ret, labels, stats, centroids = cv2.connectedComponentsWithStats(dst)
    # define the criteria to stop and refine the corners
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 
                0.001)
    corners = cv2.cornerSubPix(gray,np.float32(centroids),(5,5), 
              (-1,-1),criteria)
    
    return corners
```

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

img_six= cv2.imread("symbols/22AlphabetC.png")
img_six_gray = cv2.cvtColor(img_six,cv2.COLOR_BGR2GRAY)

img_adap_six= cv2.imread("C_01.png")
img_gray_six = cv2.cvtColor(img_adap_six,cv2.COLOR_BGR2GRAY)
img_gray_six = cv2.adaptiveThreshold(img_gray_six, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
#img_gray_six = cv2.rotate(img_gray_six, cv2.ROTATE_90_COUNTERCLOCKWISE)

sift = cv2.SIFT_create() 

kp1, des1 =sift.detectAndCompute(img_six_gray, None)
kp2, des2 = sift.detectAndCompute(img_gray_six, None)

index_params = dict(algorithm =2, trees = 10) 
search_params = dict() 
flann = cv2.FlannBasedMatcher(index_params, search_params)

matches= flann.knnMatch(des1, des2, k=2) 
  
good_points=[] 
 
for m, n in matches: 
     
    if(m.distance < 1.0*n.distance): 
        good_points.append(m) 

#img_matches = np.empty((max(img_six_gray.shape[0], img_gray_six.shape[0]), img_six_gray.shape[1]+img_gray_six.shape[1], 3), dtype=np.uint8)
#img3 = cv2.drawMatches(img_six_gray,kp1, img_gray_six,kp2,good_points,img_matches,flags=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS)

src_pts = np.float32([ kp1[m.queryIdx].pt for m in good_points]).reshape(-1,1,2)
dst_pts = np.float32([ kp2[m.trainIdx].pt for m in good_points]).reshape(-1,1,2)

M, mask = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC,5.0)
matchesMask = mask.ravel().tolist()

h, w = img_six_gray.shape

pts = np.float32([ [0,0],[0,h-1],[w-1,h-1],[w-1,0] ]).reshape(-1,1,2)
dst = cv2.perspectiveTransform(pts,M)
img_gray_six = cv2.polylines(img_gray_six,[np.int32(dst)],True,255,3, cv2.LINE_AA)

#draw the good matched key points
draw_params = dict(matchColor = (0,255,0), # draw matches in green color
                   singlePointColor = None,
                   matchesMask = matchesMask, # draw only inliers
                   flags = 2)

img3 = cv2.drawMatches(img_six_gray,kp1,img_gray_six,kp2,good_points,None,**draw_params)
plt.figure()
plt.imshow(img3, 'gray'),plt.show()
```

```python
def increase_brightness(img, value=30):
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    h, s, v = cv2.split(hsv)

    lim = 255 - value
    v[v > lim] = 255
    v[v <= lim] += value

    final_hsv = cv2.merge((h, s, v))
    img = cv2.cvtColor(final_hsv, cv2.COLOR_HSV2BGR)
    return img
```

```python
import cv2
import numpy as np

def find_corners(image_path, detected_image):
    
    if detected_image == True:
        image = cv2.imread(image_path)
        image = cv2.resize(image, [100,100])
        # image = increase_brightness(image, value=80)
        gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    else:
        image = cv2.imread(image_path)
        gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    
    gray_float = np.float32(gray)

    dst = cv2.cornerHarris(gray_float, 2, 3, 0.04)

    dst = cv2.dilate(dst, None)

    ret, dst = cv2.threshold(dst, 0.01 * dst.max(), 255, 0)
    
    cv2.imshow("dst", dst)
    
    dst = np.uint8(dst)

    # find centroids
    ret, labels, stats, centroids = cv2.connectedComponentsWithStats(dst)
    
    # define the criteria to stop and refine the corners
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 
                0.001)
    
    corners = cv2.cornerSubPix(gray,np.float32(centroids),(5,5), 
              (-1,-1),criteria)

    # To draw the corners
    for corner in corners:
        image[int(corner[1]), int(corner[0])] = [0, 0, 255]

    cv2.waitKey(0)

    return gray, corners
    
#     cv2.imshow('dst', image)
#     cv2.waitKey(0)
```

```python
import numpy as np
import math
import cv2

def find_corners(image, detected_image):
    
    hScale = 1
    wScale = 1
    if detected_image == True:
        image = cv2.imread(image)
        ih, iw, c = image.shape
        hScale = ih/100
        wScale = iw/100
        image = cv2.resize(image, [100,100])
        #image = increase_brightness(image, value=80)
        gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    else:
        image = cv2.imread(image)
        gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    
    gray_float = np.float32(gray)

    dst = cv2.cornerHarris(gray_float, 3, 9, 0.04)
    gExp = np.expand_dims(gray.copy(), -1)
    imgDraw = np.concatenate([gExp, gExp, gExp], -1)
    imgDraw[:,:,0] = dst
    
    dst = cv2.dilate(dst, None)
    
    #imgDraw[:,:,1] = dst
    cv2.imshow("dst", imgDraw)

    ret, dst = cv2.threshold(dst, 0.01 * dst.max(), 255, 0)
    
    dst = np.uint8(dst)
    
    
    # find centroids
    ret, labels, stats, centroids = cv2.connectedComponentsWithStats(dst)
    
    # define the criteria to stop and refine the corners
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 
                0.001)
    
    corners = cv2.cornerSubPix(gray,np.float32(centroids),(5,5), 
              (-1,-1),criteria)

    # # To draw the corners
    # for corner in corners:
    #     image[int(corner[1]), int(corner[0])] = [0, 0, 255]

    #corners = np.multiply(corners, )
    
    cv2.waitKey(0)
    
    return gray, corners, np.array([[wScale, hScale]])

def rotationMatrixToEulerAngles(R) :
    sy = math.sqrt(R[0,0] * R[0,0] +  R[1,0] * R[1,0])

    singular = sy < 1e-6

    if  not singular :
        x = math.atan2(R[2,1] , R[2,2])
        y = math.atan2(-R[2,0], sy)
        z = math.atan2(R[1,0], R[0,0])
    else :
        x = math.atan2(-R[1,2], R[1,1])
        y = math.atan2(-R[2,0], sy)
        z = 0

    return np.array([x, y, z])

def is_match(mapping_data, map_to_another):
    
    src_match = []
    dst_match = []
    
    for i in range(len(mapping_data)):
        for j in range(len(map_to_another)):
            if abs(mapping_data[i][0] - map_to_another[j][0]) <= 5:
                if abs(mapping_data[i][1] - map_to_another[j][1]) <= 5:
                    src_match.append(list(mapping_data[i]))
                    dst_match.append(list(map_to_another[j]))
                    break

    return src_match, dst_match

def matching_list(pts_src, pts_dst):
    
    if len(pts_src) >= len(pts_dst):
        length = len(pts_src)
        mapping_data = pts_src
        map_to_another = pts_dst
    else:
        length = len(pts_dst)
        mapping_data = pts_dst
        map_to_another = pts_src

    src_match, dst_match = is_match(mapping_data, map_to_another)
    
    return np.array(src_match), np.array(dst_match)
```

```python
def find_corner(image, detected_image):
    
    hScale = 1
    wScale = 1
    if detected_image == True:
        image = cv2.imread(image)
        ih, iw, c = image.shape
        hScale = ih/100
        wScale = iw/100
        image = cv2.resize(image, [100,100])
        #image = increase_brightness(image, value=80)
        gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    else:
        image = cv2.imread(image)
        gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    
    corners = cv2.goodFeaturesToTrack(gray, 100, 0.01, 10)
    #corners = np.int0(corners)
    
    return gray, corners, np.array([[wScale, hScale]])
```

```python
def is_match(mapping_data, map_to_another):
    
    src_match = []
    dst_match = []
    
    for i in mapping_data:
        x1,y1 = i.ravel()
        for j in map_to_another:
            x2,y2 = j.ravel()
            if abs(x2 - x1) <= 5:
                if abs(y2 - y1) <= 5:
#             if np.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2) <= 8:
                    src_match.append([x1,y1])
                    dst_match.append([x2,y2])
                    break

    return src_match, dst_match

def matching_list(pts_src, pts_dst):
    
    if len(pts_src) >= len(pts_dst):
        length = len(pts_src)
        mapping_data = pts_src
        map_to_another = pts_dst
    else:
        length = len(pts_dst)
        mapping_data = pts_dst
        map_to_another = pts_src

    src_match, dst_match = is_match(mapping_data, map_to_another)
    
    return np.array(src_match), np.array(dst_match)
```

```python
import math
#H, status = cv2.findHomography(points, objectCoord)
H, status = cv2.findHomography(dst_match, src_match)

im_out = cv2.warpPerspective(src_C, H, (100, 100))
cv2.imshow("DOG", im_out)

num, Rs, Ts, Ns  = cv2.decomposeHomographyMat(H, optimal_camera_matrix)

def rotationMatrixToEulerAngles(R) :
    sy = math.sqrt(R[0,0] * R[0,0] +  R[1,0] * R[1,0])

    singular = sy < 1e-6

    if  not singular :
        x = math.atan2(R[2,1] , R[2,2])
        y = math.atan2(-R[2,0], sy)
        z = math.atan2(R[1,0], R[0,0])
    else :
        x = math.atan2(-R[1,2], R[1,1])
        y = math.atan2(-R[2,0], sy)
        z = 0

    return np.array([x, y, z])

for rotVec in Rs:
    eulerAngle = rotationMatrixToEulerAngles(rotVec)
    print(f"Rotation Vector: \n{eulerAngle * 180/math.pi}")
cv2.waitKey(0)
```
